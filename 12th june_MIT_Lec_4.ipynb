### Guess and check algorithm

The process is called <font color='red'>exhaustive enumeration</font>

You can keep guessing untill
1. Find solution or 
2. Have guessed all values.

```python
guess = 0

x = int(input("Enter an integer: "))

while guess**2 < x:
    guess += 1
if guess ** 2 == x:
    print("Square root of", x, "is", guess)
else:
    print(x, "is not a perfect square.")
```

#### WAP that checks through all the numbers from 1 to 10 and print secret value if it's in range.
If it's not in there, it doesn't print anything.

```python
secret = int(input("Enter the number: "))
n = 0
while n <= 10:
    n += 1
    if(secret == n):
        print("Found the number after", n, "interations")
```

#### Using 'for' loop

```python
secret = 4

for i in range(1,11):
    if i == secret:
        print("Found")
```

```python
secret = 4

for i in range(1,11):
    if i == secret:
        print("Found")
        break
    else:
        print("Not found")
```
