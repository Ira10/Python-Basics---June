#### 11 . Write a function that takes two numbers as input parameters and returns true or flase depending on whether they are co-primes.

They are co-prime if they do not have any common divisor other than one.

Book - page - 83

##### 1. This code is done using math module

```python
a = int(input("Enter first number "))
b = int(input("Enter second number "))

import math

p = math.gcd(a,b)

if p == 1:
    print("Co prime")
else:
    print("No, bro")
```

```python
# Python code to demonstrate naive
# method to compute gcd ( recursion )


def hcf(a, b):
	if(b == 0):
		return a
	else:
		return hcf(b, a % b)

a = int(input("Enter first number "))
b = int(input("Enter second number "))

# prints 12
print(f"The gcd of {a} and {b} is : ", end="")

print(hcf(a,b))
```
